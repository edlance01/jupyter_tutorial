# Jupyter Notebook Quick Start

**1. The Anatomy of a Notebook**

A Jupyter Notebook is composed of cells. There are two primary types of cells:

**A. Markdown Cells (For Text & Notes)**

This cell right here is a Markdown cell. It's used for writing notes, documentation, headers, and formatted text using the simple Markdown language.

How to use:

Double-click this cell to see the raw Markdown syntax.

Press Shift + Enter (or Ctrl + Enter) to render the Markdown into formatted text.

Basic Markdown Cheatsheet:
| Syntax | Renders As |
| :--- | :--- |
| # Title | Title (H1) |
| **bold** | bold |
| *italic* | italic |
| - Item 1 | - Item 1 |
| [Link](URL) | Link |

**B. Code Cells (For Running Code)**

Code cells contain executable code (usually Python, but notebooks support many "kernels").

How to use:

Click inside a Code cell.

Write your code.

Press Shift + Enter to execute the code. The output will appear directly below the cell.

**2. Hands-On: Running Your First Code**

The cell below is a Code cell. Click on it and press Shift + Enter to run the code.

```python
# A simple Code Cell
print("Hello, Jupyter!")

# Variables persist across cells!
greeting = "Welcome to Data Science."
print(greeting)
```

Notice the In [ ]: beside the cell above. When you run it, the bracket fills with a number (In [1]). This number indicates the order of execution. If the bracket is In [*], it means the cell is currently running.

**3. Essential Tools: Data & Visualization**

Most data work involves two key libraries. You only need to import them once in any cell, and they will be available everywhere in the notebook.

**A. Pandas (Data Management)**

Used for data loading, cleaning, manipulation, and analysis. Its primary object is the DataFrame.

Run the Code Cell Below:

```python
import pandas as pd
import numpy as np

# Create a simple DataFrame (a table)
data = {
    "City": ["London", "Paris", "Berlin", "Rome"],
    "Population_M": [9.5, 2.1, 3.7, 2.8],
    "Score": np.random.randint(60, 100, 4),
}
df = pd.DataFrame(data)

# Display the DataFrame
print("--- DataFrame Preview ---")
print(df)
```

B. Matplotlib (Visualization)

Used for creating charts and graphs.

Run the Code Cell Below:

```python
import matplotlib.pyplot as plt

# Tell Jupyter to display plots inline (in the notebook)
# This is often done automatically, but good to know:
# %matplotlib inline

# Create a simple bar chart
plt.figure(figsize=(8, 4))
plt.bar(
    df["City"], df["Population_M"], color=["#4F7942", "#8FBC8F", "#A2AA8D", "#D0F0C0"]
)
plt.title("City Populations (Millions)")
plt.xlabel("City")
plt.ylabel("Population")
plt.grid(axis="y", alpha=0.5)
plt.show()
```




| Action | Keyboard Shortcut | Function |
|--------|-------------------|----------|
| Run Cell | Shift + Enter | Runs the cell, selects the next cell. |
| Run Cell | Ctrl + Enter | Runs the cell, stays on the current cell. |
| Add Cell Above | ESC then A | Switches to Command Mode (ESC) and adds a cell above. |
| Add Cell Below | ESC then B | Switches to Command Mode (ESC) and adds a cell below. |
| Change to Code | ESC then Y | Changes the selected cell type to Code. |
| Change to Markdown | ESC then M | Changes the selected cell type to Markdown. |
| Delete Cell | ESC then DD | Switches to Command Mode (ESC) and deletes the cell. |
